```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('cleaned_preprocessed_disaster_data.csv')
```

```python
df = df.drop_duplicates()
```

```python
labenc = LabelEncoder()

df['country'] = labenc.fit_transform(df['country'])
df['disaster_type'] = labenc.fit_transform(df['disaster_type'])
df['iso3'] = labenc.fit_transform(df['iso3'])
df['priority_tier'] = labenc.fit_transform(df['priority_tier'])
```

```python
scaler = StandardScaler()

numerical_cols = ['deaths','affected','mng_cap','inventory','distance','year']

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
X = df.drop('priority_tier', axis=1)
y = df['priority_tier']
```
